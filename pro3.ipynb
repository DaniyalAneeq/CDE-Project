```python
import time
import requests
from bs4 import BeautifulSoup

BASE_URL = "https://www.pakwheels.com"
all_cars = []

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

def extract_details(detail_url):
    """Extract details from car detail page."""
    try:
        r = requests.get(detail_url, headers=headers, timeout=10)
        soup = BeautifulSoup(r.text, "html.parser")

        info_ul = soup.find("ul", id="scroll_car_detail")
        details = {}

        if info_ul:
            items = info_ul.find_all("li")

            # Each pair: label -> value
            for i in range(0, len(items), 2):
                label = items[i].get_text(strip=True)
                value = items[i+1].get_text(strip=True) if i+1 < len(items) else None
                details[label] = value

        return details

    except Exception as e:
        print("Error extracting details:", e)
        return {}


for page in range(1, 11):
    print(f"Scraping page {page}...")

    url = f"{BASE_URL}/used-cars/karachi/24857?page={page}"
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    cards = soup.find_all("div", class_="col-md-9 grid-style")

    for card in cards:

        # Extract title
        title_tag = card.find("h3")
        title = title_tag.get_text(strip=True) if title_tag else None

        # Extract price
        price_tag = card.find("div", class_="price-details")
        price = price_tag.get_text(" ", strip=True) if price_tag else None

        # Extract detail page link
        a_tag = card.find("a", class_="car-name")
        if a_tag and a_tag.get("href"):
            detail_url = BASE_URL + a_tag["href"]
        else:
            detail_url = None

        # Extract summary info (year, mileage, etc.)
        ul = card.find("ul", class_="list-unstyled search-vehicle-info-2 fs13")
        li_list = [x.get_text(strip=True) for x in ul.find_all("li")] if ul else []

        # Extract detailed info from detail page
        detail_data = extract_details(detail_url) if detail_url else {}

        car_data = {
            "title": title,
            "price": price,
            "detail_url": detail_url,
            "year": li_list[0] if len(li_list) > 0 else None,
            "mileage": li_list[1] if len(li_list) > 1 else None,
            "fuel": li_list[2] if len(li_list) > 2 else None,
            "engine": li_list[3] if len(li_list) > 3 else None,
            "transmission": li_list[4] if len(li_list) > 4 else None,
            "grade": li_list[5] if len(li_list) > 5 else None,

            # Detailed page fields
            "Registered In": detail_data.get("Registered In"),
            "Color": detail_data.get("Color"),
            "Assembly": detail_data.get("Assembly"),
            "Engine Capacity": detail_data.get("Engine Capacity"),
            "Body Type": detail_data.get("Body Type"),
            "Last Updated": detail_data.get("Last Updated:"),
            "Ad Ref #": detail_data.get("Ad Ref #"),
        }

        all_cars.append(car_data)

    time.sleep(2)

print("Scraping Completed!")
print(f"Total Cars Scraped: {len(all_cars)}")
```

```python
import pandas as pd
import re

# Convert list of dicts to DataFrame
df = pd.DataFrame(all_cars)

def clean_price(price):
    """
    Convert price string into integer PKR value.
    Handles:
    - 'PKR 45.8 lacs'
    - 'PKR 3.2 million'
    - 'PKR 850,000'
    - 'Call for price'
    """

    if price is None:
        return None

    price = price.lower()

    if "call" in price:
        return None

    # Remove 'pkr', commas
    price = price.replace("pkr", "").replace(",", "").strip()

    # Handle lacs
    if "lac" in price or "lacs" in price:
        num = float(re.findall(r"[\d\.]+", price)[0])
        return int(num * 100000)

    # Handle millions
    if "million" in price:
        num = float(re.findall(r"[\d\.]+", price)[0])
        return int(num * 1000000)

    # Handle simple numbers like '850000'
    digits = re.findall(r"\d+", price)
    if digits:
        return int(digits[0])

    return None

# Apply cleaning
df["price_clean"] = df["price"].apply(clean_price)

df = df.drop(columns=["price"])
df = df.rename(columns={"price_clean": "price"})


import re

def clean_mileage(mileage):
    """
    Convert mileage like:
    - '45,000 km'
    - '92 thousand km'
    - '5.5 lac km'
    - '100 km'
    into an integer KM value.
    """

    if mileage is None:
        return None

    m = mileage.lower().strip()
    m = m.replace(",", "")

    # Remove 'km'
    m = m.replace("km", "").strip()

    # Handle thousand (92 thousand  92000)
    if "thousand" in m:
        num = float(re.findall(r"[\d\.]+", m)[0])
        return int(num * 1000)

    # Handle lac / lacs (5.5 lac  550000)
    if "lac" in m or "lacs" in m:
        num = float(re.findall(r"[\d\.]+", m)[0])
        return int(num * 100000)

    # Handle pure numeric values
    digits = re.findall(r"\d+", m)
    if digits:
        return int(digits[0])

    return None

# Apply cleaning
df["mileage_clean"] = df["mileage"].apply(clean_mileage)

df = df.drop(columns=["mileage"])
df = df.rename(columns={"mileage_clean": "mileage"})


df.to_json("data_10_pages.json", orient='records', lines=True)

df.head(10)
```
